**SIMILARLY, TABLE FOR ALL THE CUSTOMERS**

```python
customer_sales_df = (
    customer_sales_df
    .groupby('household_key')
    .agg(
        # Time difference
        time_days = ('timestamp', lambda x: (x.max() - x.min()).days),
        # No. of visits
        frequency = ('basket_id', 'nunique'),
        total_sales = ('total_sales_basket', 'sum'),
        avg_sales = ('total_sales_basket', 'mean')
    )
    .reset_index()
)
customer_sales_df
```

```python
customer_sales_df['avg_sales'].hist()
```

```python
# summary
summary_1 = ({
    # avg. sales of customer per visit
    "average_sales": customer_sales_df['avg_sales'].mean(),
    # avg. visit
    "average_purchase_frequency": customer_sales_df['frequency'].mean(),
    # churn rate
    # if the frequnecy is less than 5 then we can consider the customer to be churned
    # calculate the ratio of customer who visited store more than 5 times by dividing it by total no. of customer's visit
    # subtract it by 1 to get the rate of churned customer
    "churn_rate": 1 - (customer_sales_df['frequency'] > 5).sum() / len(customer_sales_df['frequency']),
    "max_days": customer_sales_df['time_days'].max()
}
)
summary_1_df = pd.DataFrame([summary_1])
summary_1_df
```

```python
# Define the constraints
profit_margin     = 0.15      # 15% profit on products
customer_lifetime = 5         # 5 years
eps_churn_rate    = 0.001
```

```python
# Churn CLV Calculation
summary_1_df['clv_churn_method'] = (
    summary_1_df['average_sales'] * summary_1_df['average_purchase_frequency']
    / (summary_1_df['churn_rate'] + eps_churn_rate)
) * profit_margin
```

```python
# Lifetime CLV Calculation
summary_1_df['clv_lifetime_method'] = (
    summary_1_df['average_sales'] * summary_1_df['average_purchase_frequency']
    / (summary_1_df['max_days'] / 365) * customer_lifetime
) * profit_margin
```

# 2. Cohort Model (Descriptive)

Cohort models are used to calculate the average customer lifetime value (CLV) for a group of customers or a cohort. Often times, the cohort is defined by the customer's first purchase date.

```python
# Constraints
profit_margin     = 0.15      # 15% profit on products
customer_lifetime = 5         # 5 years
eps_churn_rate    = 0.001
```

```python
# Calculate the start month for each household
df['start_month'] = (
    df
    .groupby('household_key')['timestamp']
    .transform(lambda x: x.min().strftime('%Y-%m'))
)
```

```python
df.glimpse()
```

```python
cohort_sales_df = (
    # info about each basket for a start month
    df.groupby(['start_month', 'household_key', 'basket_id'])
    .agg(
        # total sum of sales value in each basket
        total_sales_basket = ('sales_value', 'sum'),
        # timestamp of latest purchase within each basket
        # all purchase timestamps within a basket are assumed to be the same
        timestamp = ('timestamp', 'max')
    )
    .reset_index()  #  reset the multi-index to have a flat index structure
)
cohort_sales_df
```

```python
cohort_data = (
    cohort_sales_df
    .groupby(['start_month', 'household_key'])
    .agg(
        # Time difference
        time_days = ('timestamp', lambda x: (x.max() - x.min()).days),
        # No. of visits
        frequency = ('basket_id', 'nunique'),
        total_sales = ('total_sales_basket', 'sum'),
        avg_sales = ('total_sales_basket', 'mean')
    )
    .reset_index()
)
cohort_data
```

```python
# Aggregate data by start_month and household_key
cohort_data = (
    df
    .groupby(['start_month', 'household_key', 'basket_id'])
    .agg(
        total_sales_basket = ('sales_value', 'sum'),
        timestamp = ('timestamp', 'max')
    )
    .reset_index()
    .groupby(['start_month', 'household_key'])
    .agg(
        time_days = ('timestamp', lambda x: (x.max() -x.min()).days),
        frequency = ('basket_id', 'nunique'),
        total_sales=('total_sales_basket', 'sum'),
        avg_sales=('total_sales_basket', 'mean')
    ).reset_index()
)
cohort_data
```

```python
# Calculate CLV Metrics by start month
summary_2_df =  (
    cohort_data
    .groupby('start_month')
    .agg(cohort_size = ('household_key', 'nunique'),
         average_sales = ('avg_sales', 'mean'),
         average_purchase_freq = ('frequency', 'mean'),
         churn_rate = ('frequency', lambda x: 1-(x > 5).sum()/len(x)),
         max_days = ('time_days', 'max')
        ).reset_index()
)
summary_2_df
```

```python
cohort_data[cohort_data['start_month'] < '2017-05']
```

```python
# Calculate CLV Metrics by start month
summary_2_df_clear =  (
    cohort_data[cohort_data['start_month'] < '2017-05'] # we will trim later cohorts due to their small size
    .groupby('start_month')
    .agg(cohort_size = ('household_key', 'nunique'),
         average_sales = ('avg_sales', 'mean'),
         average_purchase_freq = ('frequency', 'mean'),
         churn_rate = ('frequency', lambda x: 1-(x > 5).sum()/len(x)),
         max_days = ('time_days', 'max')
        ).reset_index()
)
summary_2_df_clear
```

```python
# Churn CLV Calculation
summary_2_df_clear['clv_churn_method'] = (
    summary_2_df_clear['average_sales'] * summary_2_df_clear['average_purchase_freq']
    / (summary_2_df_clear['churn_rate'] + eps_churn_rate)
) * profit_margin
```

```python
# Lifetime CLV Calculation
summary_2_df_clear['clv_lifetime_method'] = (
    summary_2_df_clear['average_sales'] * summary_2_df_clear['average_purchase_freq']
    / (summary_2_df_clear['max_days'] / 365) * customer_lifetime
) * profit_margin
```

```python
summary_2_df_clear = summary_2_df_clear.set_index('start_month')
summary_2_df_clear
```

**Conclusion:**
*   I don't trust these CLV Calculations. They are super optimistic for the churn calculations. Lifetime is a bit more realistic, but still has not earned my trust. Churn rates can be useful for subscription based companies (e-commerce)
*   The bottom line is that as we get more granular, we can gain higher accuracy in our CLV calculations.
*   The next step is to build predictive models to forecast future CLV

**RFM (Recenecy, Frequency, Monetary Value) Table**
> It will calculate the following for each customer
* Frequency: How many transactions
* Recency: How recent was their last transaction
* T: Total purchase
* monetary_value: mean spend

```python
# Probabilistic Models are used to predict the future transactions and churn rate of a customer

# Data Preparation
summary_3_df = lf.utils.summary_data_from_transaction_data(
    df,
    customer_id_col = 'household_key',
    datetime_col = 'timestamp',
    monetary_value_col = 'sales_value'
)
summary_3_df
```

```python
# Removing customers that have 0 or negative values
# will get an error in Gamma-Gamma model if there are non-positive values

summary_3_df[summary_3_df['monetary_value'] <= 0]
```

```python
summary_3_df = summary_3_df[summary_3_df['monetary_value'] > 0]
```

**MODEL 1: BG/NBD model -----**

* The BG/NBD model is a probabilistic model that predicts the number of repeat purchases a customer will make
* The model can only predict the future transactions and churn rate of a customer

```python
# NOTE: It doesn't converge, increase penalizer_coef
bgf = lf.BetaGeoFitter(penalizer_coef=0.15)
```

```python
bgf.fit(summary_3_df['frequency'], summary_3_df['recency'], summary_3_df['T'])
```

```python
bgf.summary
```

```python
# Conditional Probability alive
summary_3_df['probability_alive'] = bgf.conditional_probability_alive(
    summary_3_df['frequency'], summary_3_df['recency'], summary_3_df['T'])
```

```python
summary_3_df
```

```python
# Visualizing the probability alive matrix
plot_probability_alive_matrix(bgf, cmap = 'viridis')
```

* If the Customer's Recency = 600 then they have to purchase at least 50 times to be alive.
* If a customer has bought multiple times (frequency) and the time between the first & last transaction is high (recency), then his/her probability of being alive is high
* If a customer has less frequency (bought once or twice) and the time between first & last transaction is low (recency), then his/her probability of being alive is low

**MODEL 2: GAMMA-GAMMA MODEL ----**
> It is used to predict the average transcations value for each customer

```python
ggf = lf.GammaGammaFitter(penalizer_coef=0.1)
```

```python
ggf.fit(summary_3_df['frequency'], summary_3_df['monetary_value'])
```

```python
summary_3_df['predicted_avg_sales'] = ggf.conditional_expected_average_profit(
    summary_3_df['frequency'],
    summary_3_df['monetary_value']
)
```

```python
summary_3_df
```

```python
# Predict the Customer Lifetime Value for Next 90 Days ----

summary_3_df['predicted_clv_3mon'] = ggf.customer_lifetime_value(
    bgf,
    summary_3_df['frequency'],
    summary_3_df['recency'],
    summary_3_df['T'],
    summary_3_df['monetary_value'],
    time=3,
    freq='D',
    discount_rate=0.01    # 1% discount rate
)
```

```python
summary_3_df
```

```python
# Don't forget to add in profit margin
summary_3_df['predicted_profit_3mon'] = summary_3_df['predicted_clv_3mon'] * profit_margin
```

```python
summary_3_df
```

CONCLUSIONS ----
* Now we know how much we can spend to target these customers in the next 3 months (predicted_profit_3mon)
* Much better than historical/descriptive models as it can predict the future transactionsand churn rate of a customer
* However, we don't know what relationships exist between the features and the lifetime value of a customer, and therefore, we can't confidently improve the business
* The next step is to build predictive models to forecast future customer lifetime value AND to understand the relationships between the features and the lifetime value of a customer.

# 4. Machine Learning Model

> Framing the problem

```python
# LIBRARIES
import pandas as pd
import pytimetk as tk
import pycaret.classification as clf
import pycaret.regression as reg
```

```python
# CONSTRAINTS
profit_margin = 0.15    # 15% profit on products
```

```python
# DATA
transactions_df = pd.read_csv('/content/drive/MyDrive/CLV Project/transactions_processed.csv')
```

```python
df = transactions_df.copy()
```

```python
df.glimpse()
```

```python
n_days = 90
max_date = df['timestamp'].max()
cutoff = max_date - pd.to_timedelta(n_days, unit = 'd')
```

```python
# Train-Test Split
temporal_in_df = df[df['timestamp'] < cutoff]
```

```python
temporal_out_df = (
    df[df['timestamp'] >= cutoff]
    .query('household_key in @temporal_in_df.household_key')
)
temporal_out_df
```

Feature Engineeing:
> 2-Stage Process
* Need to frame the problem
* Need to think about what features to include

```python
# Target from the out data (90 days data)
target_df = (
    temporal_out_df[['household_key', 'timestamp', 'sales_value']]
    .groupby('household_key')
    .sum()
    .rename({'sales_value': 'sales_90_value'}, axis = 1)
    .assign(sales_90_flag = 1)
)
target_df
```

```python
target_df['sales_90_flag'].value_counts()
```

```python
max_date = temporal_in_df['timestamp'].max()
max_date
```

Let's create three auxiliary dataframes:

* recency_features_df - contains the measure of purchase recency
* frequency_features_df - contains the number of purchases made by the customer
* monetary_features_df - contains the purchase volume of the customer and the average order size

```python
# Make Recency(DATE) Features from in data
recency_features_df = (
    temporal_in_df[['household_key', 'timestamp']]
    .groupby('household_key')
    .apply(
        lambda x: int((x['timestamp'].max() - max_date) / pd.to_timedelta(1, 'day'))
    )
    .to_frame()
    .set_axis(['recency'], axis = 1)
)
recency_features_df
```

```python
# Make Frequency(COUNT) Features from in data
frequency_features_df = (
    temporal_in_df[['household_key', 'timestamp']]
    .groupby('household_key')
    .count()
    .set_axis(['frequency'], axis=1)
)
frequency_features_df
```

```python
# Make Price(MONETARY) Features from in data
monetary_features_df = (
    temporal_in_df
    .groupby('household_key')
    .aggregate({'sales_value': ['sum', 'mean']})
    .set_axis(['sales_value_sum', 'sales_value_mmean'], axis = 1)
)
monetary_features_df
```

```python
# Transactions Last Month
cutoff_28d = cutoff - pd.to_timedelta(28, unit='d')
print(cutoff_28d)

transactions_last_month_df = (
    temporal_in_df[['household_key', 'timestamp']]
    .drop_duplicates()
    .query('timestamp >= @cutoff_28d')
    .groupby('household_key')
    .size()
    .to_frame()
    .set_axis(['transactions_last_month'], axis=1)
)
transactions_last_month_df
```

```python
# Transactions Last 2 Weeks
cutoff_14d = cutoff - pd.to_timedelta(14, unit='d')
print(cutoff_14d)

transactions_last_2weeks_df = (
    temporal_in_df[['household_key', 'timestamp']]
    .drop_duplicates()
    .query('timestamp >= @cutoff_14d')
    .groupby('household_key')
    .size()
    .to_frame()
    .set_axis(['transactions_last_2weeks'], axis=1)
)
transactions_last_2weeks_df
```

```python
# Spend Last 2 Weeks
cutoff_14d = cutoff - pd.to_timedelta(14, unit='d')
print(cutoff_14d)

sales_last_2weeks_df = (
    temporal_in_df[['household_key', 'timestamp', 'sales_value']]
    .drop_duplicates()
    .query('timestamp >= @cutoff_14d')
    .groupby('household_key')
    .sum()
    .set_axis(['sales_value_last_2weeks'], axis=1)
)
sales_last_2weeks_df
```

```python
# COMBINING FEATURES
features_df = (
    pd.concat(
        [recency_features_df,
         frequency_features_df,
         monetary_features_df,
         transactions_last_month_df,
         transactions_last_2weeks_df,
         spend_last_2weeks_df],
        axis=1)
    .merge(
        target_df,
        left_index = True,
        right_index = True,
        how = 'left'
    ).fillna(0)
)
features_df
```

```python
features_df['sales_90_flag'].value_counts()
```

Machine Learning -----
> Will use pycaret to quickly build a predictive model

```python
# REGRESSION
reg_setup = reg.setup(
    data = features_df.drop('sales_90_flag', axis = 1),
    target = 'sales_90_value',
    train_size = 0.8,
    normalize = True,
    session_id = 123,
    verbose = True,
    log_experiment = False
)

xgb_reg_model = reg.create_model('xgboost')
reg_predictions_df = (
    reg.predict_model(xgb_reg_model, data = features_df)
    .sort_values('prediction_label', ascending = False)
)

reg_predictions_df
```

```python
# CLASSIFICATION (SPEND PROBABILITY) ----

clf_setup = clf.setup(
    data = features_df.drop('sales_90_value', axis=1),
    target = 'sales_90_flag',
    train_size = 0.8,
    normalize = True,
    session_id = 123,
    verbose = True,
    log_experiment = False
)

xgb_clf_model = clf.create_model('xgboost')

clf_predictions_df = (
    clf.predict_model(xgb_clf_model, data = features_df, raw_score = True)
    .sort_values('prediction_score_1', ascending = False)
)

clf_predictions_df
```
